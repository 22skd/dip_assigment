```python
import cv2
import numpy as np

# Read an image from file in grayscale
image = cv2.imread('/content/image1.jpeg')

# Perform 2-D DFT
dft_image = np.fft.fft2(image)

# Shift the zero frequency component to the center of the spectrum
shifted_dft = np.fft.fftshift(dft_image)

# Display the magnitude spectrum
magnitude_spectrum = 20 * np.log(np.abs(shifted_dft))
print('Magnitude Spectrum')
cv2_imshow( magnitude_spectrum)
cv2.waitKey(0)
cv2.destroyAllWindows()




```
