```python
import cv2
import numpy as np

# Read an image from file in grayscale
image = cv2.imread('/content/image2.jpeg', 0)

# Perform erosion and dilation operations
kernel = np.ones((5, 5), np.uint8)
eroded_image = cv2.erode(image, kernel, iterations=1)
dilated_image = cv2.dilate(image, kernel, iterations=1)

# Display the eroded and dilated images
print('Eroded Image')
cv2_imshow( eroded_image)
print('dilated Image')
cv2_imshow( dilated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
