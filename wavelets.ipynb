```python
import cv2
import pywt

# Read an image from file
image = cv2.imread('/content/image1.jpeg')

# Apply wavelet transformation using the Haar wavelet
coeffs = pywt.dwt2(image, 'haar')

# Reconstruct the image from the wavelet coefficients
reconstructed_image = pywt.idwt2(coeffs, 'haar')

# Display the original and reconstructed images
print('Original Image')
cv2_imshow( image)
print('Reconstructed Image')
cv2_imshow( reconstructed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()



```
