```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Read an image from file
image = cv2.imread('/content/image2.jpeg', 0)

# Apply the Butterworth lowpass filter in the frequency domain
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
mask = np.zeros((rows, cols), np.uint8)
r = 50  # Radius of the lowpass filter
n = 2   # Order of the Butterworth filter
mask[crow - r:crow + r, ccol - r:ccol + r] = 1
filtered_image = np.fft.ifftshift(np.fft.ifft2(np.fft.fftshift(np.fft.fft2(image)) * mask))

# Display the filtered image
print('Filtered Image')
cv2_imshow( np.abs(filtered_image))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
